## How much money each state spend on propositions?

1- Importing the pandas lib and configuring the display format of decimal values.

```python
import pandas as pd
pd.options.display.float_format = '${:,.2f}'.format
```

2- Reading the CSV files.

```python
comm = pd.read_csv("./committees.csv")
cont = pd.read_csv("./contributions.csv")
```

3- Calculating the total of contributors by each state.

```python
state_person_total = cont.groupby("contributor_state").size().reset_index(name="total_contributors")
```

4- Calculating the total of money spend by each state.

```python
state_person_cont = cont.groupby("contributor_state").sum()["amount"].reset_index().sort_values(by="amount", ascending=False)
```

5- Merging the total of contributors and total of money

```python
state_data = state_person_cont.merge(state_person_total, on="contributor_state")
```

6- Calculating statistic: the amount spend by each contributor on each state.

```python
amount_contributor = pd.DataFrame.from_dict({"amount per contributor": state_data.amount / state_data.total_contributors, "contributor_state": state_data.contributor_state})
```

7- Merging all data together to show the final result.

```python
state_data.merge(amount_contributor, on="contributor_state")
```
